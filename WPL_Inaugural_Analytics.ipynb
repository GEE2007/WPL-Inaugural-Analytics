#### Chart - 4

````python
# Chart - 4 visualization code
economy = bowling.groupby('bowlerName').agg({
    'economy': 'mean',
    'overs': 'sum'
})
economy = economy[economy['overs'] >= 10]
economy = economy.sort_values('economy').head(10)
economy['economy'].plot(kind='bar', figsize=(9,3))
plt.title("Top 10 Most Economical Bowlers")
plt.xlabel("Bowlers")
plt.ylabel("Economy Rate")
plt.show()
````

#### Chart - 7

````python
# Chart - 7 visualization code
batting['SR'] = pd.to_numeric(batting['SR'], errors='coerce')

player_stats = batting.groupby('battername').agg({
    'runs':'sum',
    'SR':'mean'
})
plt.figure(figsize=(8,5))
plt.scatter(player_stats['SR'], player_stats['runs'])
z = np.polyfit(player_stats['SR'], player_stats['runs'], 1)
p = np.poly1d(z)
plt.plot(player_stats['SR'], p(player_stats['SR']))

plt.title("Average Strike Rate vs Total Runs")
plt.xlabel("Average Strike Rate")
plt.ylabel("Total Runs")
plt.grid(True)

plt.show()
````

##### 1. Why did you pick the specific chart?

*A scatter plot was chosen because it helps visualize the relationship between a batter's average strike rate and the total runs scored.*

*   *It helps identify trends, outliers, and different batting patterns among players.*

##### 2. What is/are the insight(s) found from the chart?



*   *A positive relationship can be observed between strike rate and total runs, indicating that many high-scoring batters also maintained healthy strike rates.*


*   *A few players achieved high strike rates but scored comparatively fewer runs, suggesting aggressive batting in limited opportunities.*

*   *Some players accumulated more runs with moderate strike rates, reflecting consistent innings-building rather than explosive scoring.*






##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

**Positive Business Impact** :

*   *This analysis helps batting coaches and team analysts identify different batting styles, such as aggressive finishers and consistent run scorers.*
*   *It can assist in deciding batting order, player roles, and match strategies based on scoring patterns.*

**Negative Growth** :


*   *Players with both low strike rates and low run totals may reduce the team's scoring momentum in the T20 format.*
*   *If multiple players display similar performances, the team may struggle to post or chase competitive totals.*
