#### Chart - 2

````python
# Chart - 2 visualization code
top_runs = batting.groupby('battername')['runs'].sum().sort_values(ascending=False).head(10)

top_runs.plot(kind='bar', figsize=(9,3))

plt.title("Top 10 Run Scorers in WPL")
plt.xlabel("Players")
plt.ylabel("Total Runs")

plt.show()
````

##### 1. Why did you pick the specific chart?



*   *A bar chart is ideal for comparing the total runs scored by different batters.*
*   *It allows quick identification of the highest run scorers and makes ranking players simple and visually clear.*



##### 2. What is/are the insight(s) found from the chart?



*   *Meg Lanning secured the Orange Cap with 345 runs, and her batting performances played a significant role in helping Delhi Capitals finish as runners-up.*



*   *The chart identifies the tournament's highest run scorers and highlights players who made significant batting contributions.*



##### 3. Will the gained insights help creating a positive business impact?
Are there any insights that lead to negative growth? Justify with specific reason.

**Positive Business Impact:**


*   *It helps team management, franchises, brands, and sponsors identify high-performing batters.*

#### Chart - 3

````python
# Chart - 3 visualization code
top_wickets = bowling.groupby('bowlerName')['wickets'].sum().sort_values(ascending=False).head(10)
top_wickets.plot(kind='bar', figsize=(9,3))
plt.title("Top 10 Wicket Takers in WPL")
plt.xlabel("Bowlers")
plt.ylabel("Total Wickets")
plt.show()
````

##### 1. Why did you pick the specific chart?

*I chose a bar chart because it makes it easy to compare the total wickets taken by different bowlers.*

##### 2. What is/are the insight(s) found from the chart?
